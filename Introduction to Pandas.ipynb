#### Let's summarise the missing values with custom functions

<h5>Presenting missing values in a table form</h5>

```python
def missing_data(df):
    total = df.isnull().sum()
    percent = (df.isnull().sum()/df.isnull().count()*100)
    missing_values = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    types = []
    for col in df.columns:
        dtype = str(df[col].dtype)
        types.append(dtype)
    missing_values['Types'] = types
    missing_values.sort_values('Total',ascending=False,inplace=True)
    return missing_values.T

missing_data(df)
```

##### Plotting missing values with `matplotlib`

```python
import matplotlib.pyplot as plt
```

```python

def plot_missing_values(df):
    total = df.isnull().sum()
    percent = (df.isnull().sum()/df.isnull().count()*100)
    missing_values = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    missing_values.sort_values('Total', ascending=False, inplace=True)
    
    f, ax = plt.subplots(figsize=(15, 6))
    plt.xticks(rotation='0')
    plt.bar(x=missing_values.index, height=missing_values['Percent'], color='navy')
    plt.xlabel('Column Heads', fontsize=15)
    plt.ylabel('Percentage of missing values', fontsize=15)
    plt.title('Percentage of missing data by feature', fontsize=15)
    plt.grid(which='both', color='grey', linestyle='-', linewidth=1, alpha=0.7)
    return plt.show()

plot_missing_values(df)
```

    <li><b>df.fillna(axis=0, method='ffill')</b>. when the <b>method</b> parameter is set to <b>'ffill'</b>, it replaces missing values by propagating the last valid observation forward to next valid observation in <b>axis=0.</b></li>
    <li><b>df.fillna(axis=0, method='bfill')</b>. when the <b>method</b> parameter is set to <b>'bfill'</b>, it replaces missing values by using the next valid observation to fill gap in <b>axis=0.</b></li>
</ul>
</div>

### Apart from `.fillna()`, missing values can also be handled using `.dropna()`.

<div class="alert alert-block alert-info">
<b>Tip!:</b> <b>.dropna()</b> drops rows and columns with missing values.
    It has a parameter <b>how</b> that is used to set the condition to drop rows or columns with missing values.
<ul>
    <li><b>how='any'</b> drops a particular row or column if there is any missing value.</li>
    <li><b>how='all'</b> drops a particular row or column if all values are missing.</li>
</ul>
    <p> It also have a parameter called <b>thresh</b> which when set, drops a column or a row base on the threshold value.
     <p> And an <b>axis</b> parameter which is also used to select the <b>row (0)</b> or <b>column (1)</b> we want to drop missing values from.   
</div>


#### `df.dropna(how='all', axis=0, inplace=True)`
<i>Setting the <b>how</b> parameter to 'all' drops axis=0 if all values missing</i>

#### `df.dropna(thresh=3, axis=0, inplace=True)`
<i>Setting the <b>thresh</b> parameter to 3 drops axis=0 with at least 3 missing values</i>
